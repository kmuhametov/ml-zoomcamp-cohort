```python
import pandas as pd
import numpy as np
from sklearn.feature_selection import chi2
```

```python
data = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df.head()
```

```python
df.T
```

```python
df['Attrition'] = df['Attrition'].map({'Yes': 1, 'No': 0})
```

```python
categorical_features = df.select_dtypes(include=['object', 'category']).columns.tolist()
numerical_features = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
```

```python
print("Categorical:", categorical_features)
print("Numerical:", numerical_features)
```

```python
df[categorical_features] = df[categorical_features].fillna('NA')
df[numerical_features] = df[numerical_features].fillna(0)
```

```python
df.describe()
```

```python
corr = df.corr(numeric_only=True)['Attrition'].abs().sort_values(ascending=False)
corr
```

```python
#delete feateres with correlation below 0.10
```

```python
drop_cols = ['DistanceFromHome', 'WorkLifeBalance', 'TrainingTimesLastYear', 'DailyRate', 'RelationshipSatisfaction', 'NumCompaniesWorked', 'YearsSinceLastPromotion', 'Education', 'MonthlyRate', 'PercentSalaryHike', 'EmployeeNumber', 'HourlyRate', 'PerformanceRating', 'EmployeeCount', 'StandardHours']
df = df.drop(columns=drop_cols)
```

```python
df_categorical_features = df[categorical_features]
```

```python
df_categorical_features
```

```python
df_categorical_features = df[categorical_features]

X = pd.get_dummies(df_categorical_features)
y = df['Attrition']

chi_scores, p_values = chi2(X, y)

results = pd.DataFrame({
    'feature': X.columns,
    'chi2': chi_scores,
    'p_value': p_values
}).sort_values('chi2', ascending=False)
pd.set_option('display.float_format', '{:.10f}'.format)
print(results)
```

```python
drop_cols_categorical = ['Over18', 'Gender']
df = df.drop(columns=drop_cols_categorical)
```
